```python
import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
import pickle
version_info = {'scikit-learn': sklearn.__version__}
```

```python
##load dataset
df=pd.read_csv('Churn_Modelling.csv')
df
```

```python
df.isna().sum()
```

### Preprocessing

```python
## drop unnecessary columns
df.drop(['RowNumber','CustomerId','Surname'],axis=1,inplace=True)
df
```

```python
## encode categorical variabels
gender_encoder=LabelEncoder()
df['Gender']=gender_encoder.fit_transform(df['Gender'])
```

```python
df
```

```python
## OHE for geography column

from sklearn.preprocessing import OneHotEncoder

ohe=OneHotEncoder()
geo_encoded=ohe.fit_transform(df[['Geography']])
geo_encoded
```

```python
ohe.get_feature_names_out(['Geography'])
```

```python
geo_encoded_df=pd.DataFrame(geo_encoded.toarray(),columns=ohe.get_feature_names_out(['Geography']))
```

```python
geo_encoded_df
```

```python
df
```

```python
### combine OHE columns with df

df=pd.concat([df.drop('Geography',axis=1),geo_encoded_df],axis=1)
```

```python
df
```

```python
### save label encoder and OHE into pickle file
with open('gender_encoder.pkl','wb') as file:
    pickle.dump((gender_encoder,version_info),file)

with open('ohe_geo.pkl','wb') as file:
    pickle.dump((ohe,version_info),file)
```

```python
### divide the dataset into dependent and independent features

y=df['Exited']
X=df.drop('Exited',axis=1)
```

```python
## splitting X and y
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
# scale the features
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
## dump scaler into pickle file
with open('scaler.pkl','wb') as file:
    pickle.dump((scaler,version_info),file)
```

### ANN implementation

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
```

```python
model=Sequential([
    Dense(64,activation='relu',input_shape=(X_train.shape[1],)),## first hiddenlayer with input
    Dense(32,activation='relu'),##HL2
    Dense(1,activation='sigmoid')##output
]
    
)
```

```python
model.summary()
```

```python
## compile the model(to forward and backward propogation)
opt=tf.keras.optimizers.Adam(learning_rate=0.01) ##customized learning rate
model.compile(optimizer=opt,loss='binary_crossentropy',metrics=['accuracy'])## uses default learning rate for optimizer if directly mention in compile
```

```python
## set up the tensorboard

log_dir='logs/fot'+datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tf_callback=TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
## set up early stopping
## used when no improvements over the epochs
early_stopping=EarlyStopping(monitor='val_loss',patience=15,restore_best_weights=True)
```

```python
## train the model
history=model.fit(
    X_train,y_train,validation_data=(X_test,y_test),epochs=100,
    callbacks=[tf_callback,early_stopping]
    )
```

```python
#save model
model.save('model.h5')
```
